```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
audusd = pd.read_csv("aud.csv")
eurusd = pd.read_csv("eur.csv")
gbpusd = pd.read_csv("gbp.csv")
nzdusd = pd.read_csv("nzd.csv")
usdcad = pd.read_csv("cad.csv")
usdchf = pd.read_csv("chf.csv")
usdjpy = pd.read_csv("jpy.csv")
```

```python
eurusd.head()
```

```python
audusd["Local time"] = pd.to_datetime(audusd["Local time"])
eurusd["Local time"] = pd.to_datetime(eurusd["Local time"])
nzdusd["Local time"] = pd.to_datetime(nzdusd["Local time"])
gbpusd["Local time"] = pd.to_datetime(gbpusd["Local time"])
usdcad["Local time"] = pd.to_datetime(usdcad["Local time"])
usdchf["Local time"] = pd.to_datetime(usdchf["Local time"])
usdjpy["Local time"] = pd.to_datetime(usdjpy["Local time"])
```

```python
audusd["year"] = audusd["Local time"].dt.year
eurusd["year"] = eurusd["Local time"].dt.year
nzdusd["year"] = nzdusd["Local time"].dt.year
gbpusd["year"] = gbpusd["Local time"].dt.year
usdcad["year"] = usdcad["Local time"].dt.year
usdchf["year"] = usdchf["Local time"].dt.year
usdjpy["year"] = usdjpy["Local time"].dt.year
```

```python
audusd.head()
```

```python
audusdclean = audusd[audusd['Volume']!=0 ].reset_index(drop=True)
eurusdclean = eurusd[eurusd['Volume']!=0 ].reset_index(drop=True)
gbpusdclean = gbpusd[gbpusd['Volume']!=0 ].reset_index(drop=True)
nzdusdclean = nzdusd[nzdusd['Volume']!=0 ].reset_index(drop=True)
usdcadclean = usdcad[usdcad['Volume']!=0 ].reset_index(drop=True)
usdchfclean = usdchf[usdchf['Volume']!=0 ].reset_index(drop=True)
usdjpyclean = usdjpy[usdjpy['Volume']!=0 ].reset_index(drop=True)
```

```python
audusddf = audusdclean['Close']
eurusddf = eurusdclean['Close']
gbpusddf = gbpusdclean['Close']
nzdusddf = nzdusdclean['Close']
usdcaddf = usdcadclean['Close']
usdchfdf = usdchfclean['Close']
usdjpydf = usdjpyclean['Close']
```
