## Feature engeneering

```python
data2=data.copy()
```

```python
#creating year joined column
data2['Year_joined']=data2['Joined'].apply(lambda x:x.split(',')[1])
```

```python
#removing '\n'
data2['Team_contract']=data2['Team_contract'].apply(lambda x: x.replace('\n',''))
```

```python
#some players are on loan so they have loan contract
data2['Team_contract'].unique()
```

```python
#only players on loan have their loan contracts either ending on June30 2021 or jun30 2022
#Identifying players on loan
on_loan=data2[(data2['Team_contract']=='Jun 30, 2021 On Loan') | (data2['Team_contract']=='Jun 30, 2022 On Loan')]
on_loan
```

```python
#Creating an on loan column
data2['On_loan']=data2['Team_contract'].apply(lambda x:1 if 'on loan'in x.lower() else 0)
```

```python
def impute_contract_end(cols):
    '''this function slices the end of contract dates'''
    team_contract=cols[0]
    on_loan=cols[1]
    if on_loan==0:
        return team_contract[7:12]
    else:
        return team_contract[8:12]
                                                          
```

```python
#creating endcontract column
data2['Contract_end']=data2[['Team_contract','On_loan']].apply(impute_contract_end,axis=1)
```

```python
#changing height and weight from imperial to metric conversion system
#(feet,inches to meters,centimeters)
def imperial_height_converter(height):
    height=height.replace('"','')
    return round(((float(height[0])*12) + float(height[2:]))*0.0254,2)

#(pounds to kgs)
def imperial_weight_converter(weight):
    weight=weight.replace('lbs','')
    return round(float(weight)*0.453592)
```

```python
data2['Height']=data2['Height'].apply(imperial_height_converter)
data2['Weight']=data2['Weight'].apply(imperial_weight_converter)
```

```python
#Cleaning international_rating column
data2['International_rating']=data2['International_rating'].apply(lambda x:x.split(' ')[0])
```

```python
#converting wage columns to float 
def wage_converter(wage):
    
    if 'K' in wage:
        #removing euro sign by slicing
        num1=wage[1:].replace('K','')
        return float(num1)*1000
    else:
        num1=wage[1:]
        return (float(num1))

#converting value to float
def value_converter(value):
    
    if 'M' in value:
        #removing euro sign by slicing
        num2=value[1:].replace('M','')
        return float(num2)*1000000
    elif 'K' in value:
        num2=value[1:].replace('K','')
        return float(num2)*1000
    else:
        num2=value[1:]
        return (float(num2))
    
```

```python
data2['Value']=data2['Value'].apply(value_converter)
data2['Wage']=data2['Wage'].apply(wage_converter)
```

```python
data2['Current_year']=2021
```

```python
#creating new column
data2['Remaining_contract_years']=data2['Contract_end'].astype('float')-data2['Current_year']
```

```python
def years_spent_in_club(cols):
    '''this function calculates years spent on a club given a player is on loan or not'''
    year_joined=cols[0]
    on_loan=cols[1]
    current_year=cols[2]
    if on_loan==0:
        return current_year-float(year_joined)
    else:
        #On average players on loan last one season or less,some of them are not even from the English Premier League
        #So i will use 1 year as the duration of those players
        return 1
```

```python
data2['Years_in_the_club']=data2[['Year_joined','On_loan','Current_year']].apply(years_spent_in_club,axis=1)
```

```python
#splitting the data to train test split before further data preprocessing to avoid data leakage
from sklearn.model_selection import train_test_split
```

```python
x=data2.drop('Wage',axis=1)
y=data2['Wage']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#Joining x-train and y_train
Train_data = (x_train.join(y_train)).reset_index(drop=True)
Train_data.head()
```

```python
#Joining x-test and y-test
Test_data=(x_test.join(y_test)).reset_index(drop=True)
Test_data.head()
```
